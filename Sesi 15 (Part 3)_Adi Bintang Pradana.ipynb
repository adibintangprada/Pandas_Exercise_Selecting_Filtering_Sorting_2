## COVID-19

Memuat dan mencetak informasi data set dasar:

```python
import pandas as pd
covid = pd.read_csv('covid19.csv')
covid
```

```python
#data set head
covid.head()
```

```python
#dataset tail
covid.tail()
```

```python
# hitung berapa banyak missing values di setiap kolomnya
covid.isnull().sum()
```

```python
covid = covid.fillna(0).astype({'Casos':'int', 'Hospitalizados':'int', 'UCI':'int', 'Fallecidos':'int', 'Recuperados':'int'})
covid
```

### Menulis program Python untuk mendapatkan jumlah kasus yang dikonfirmasi, dirawat di rumah sakit, dipindahkan ke ICU, meninggal dan sembuh.

```python
covid[['Casos', 'Hospitalizados', 'UCI', 'Fallecidos', 'Recuperados']].sum().to_frame().transpose()
```

### Temukan jumlah terakhir dari kasus yang dikonfirmasi, meninggal dan dipulihkan oleh Komunitas:

```python
covid[['CCAA', 'Casos', 'Fallecidos', 'Recuperados']].groupby(['CCAA']).sum().sort_values(by=(['Casos', 'Fallecidos', 'Recuperados']), ascending = False)
```

### Temukan jumlah kasus terbaru yang dikonfirmasi, dirawat di rumah sakit, dipindahkan ke ICU, meninggal dan pulih dari COVID-19 di Navarra

```python
covid[covid['CCAA'] == 'Navarra'].sort_values(by = ['Fecha'], ascending = False).head(1)
```

### Temukan kolom baru dengan kasus aktif:

```python
covid['Activos'] = covid['Casos'] + (covid['Recuperados'] - covid['Fallecidos'])
covid
```

### Temukan kapan kasus COVID-19 pertama dan kematian terjadi di Spanyol

```python
# cari kasus pertama (casos)
covid[covid['Casos'] == 1].sort_values(by = 'Fecha', ascending = True).head(1)
```

```python
# selediki kematian pertama ( falecidos )
covid[covid['Fallecidos'] == 1].sort_values(by = 'Fecha', ascending = True).head(1)
```

### Menemukan kapan kematian pertama kali terjadi di setiap Komunitas

```python
covid[covid['Fallecidos'] > 0].sort_values(by = 'Fecha', ascending = True).groupby('CCAA').head(1)
```
